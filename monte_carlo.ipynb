```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def PiMonteCarlo(n_iter):
    pi_list = []
    for i in range(1000):
        points_in = []
        points_out = []
        is_in = 0
        for i in range(n_iter):
            rand_x, rand_y= (random.random()*2-1,  random.random()*2-1)
            point = [rand_x,rand_y]
            dist = np.linalg.norm(point)
            if dist < 1:
                is_in +=1
                points_in.append(point)
            else:
                points_out.append(point)
            pi_n = is_in/n_iter*4
        pi_list.append(pi_n)

    x_in = [point[0] for point in points_in]
    y_in = [point[1] for point in points_in]

    y_out = [point[1] for point in points_out]
    x_out = [point[0] for point in points_out]
    return [x_in, y_in] , [x_out,y_out], pi_list
```

```python
def xy(r,phi):
  return r*np.cos(phi), r*np.sin(phi)
```

```python
def PlotMonteCarlo(n_iter, type):
    fig = plt.figure()
    p_in , p_out, pi_list = PiMonteCarlo(n_iter)
    if(type == "hist"):
        n, bins, patches = plt.hist(x=pi_list, bins='auto', color='#0504aa',
                                alpha=0.7)

        plt.grid(axis='y', alpha=0.75)
        plt.xlabel('Value')
        plt.ylabel('Frequency')
        plt.title(r'Pi distribution: $\mu={:.4}$'.format(np.mean(pi_list)))

    else:
        ax = fig.add_subplot(111,aspect='equal')  

        phis=np.arange(0,6.28,0.01)
        r =1.
        ax.plot( *xy(r,phis), c='r',ls='-' )
        plt.scatter(p_in[0],p_in[1], color="r")
        plt.scatter(p_out[0],p_out[1], color="k")
        plt.xlim=[-1,1]
        plt.ylim=[-1,1]
    plt.show()    
    return p_in , p_out, pi_list
```

```python
PlotMonteCarlo(50,"hist" );
```

```python
PlotMonteCarlo(100, "hist");
```
